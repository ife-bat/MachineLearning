# ML on IFE data

**In this notebook:**
- A dictionary is created to make it easy to access all data
    - One for all cells and others for the subsets of cells
- A simple Elastic Net model is created
    - Cycle life defined as the normalized cycle number when normalized capacity reaches 90% (first c20 cycle when this occurs)
    - Features are generated from the difference between the first and last C20 cycle, common name c20_diff
        - variance and mean of Qd_c20_diff, dQdV_c20_diff, total 4 features
- Next step will be to
    - Seperate cells with similar cycling conditions (Concern: too few samples)
    - Include more features
    - Find out how to make model with the rest of the cycling data

### Extract relevant data

- f1: $Q_{[100,230] - first}$ for C20 cycles
- f2: $dQdV_{[100,230] - first}$ for C20 cycles
- f3: $Q_{115-15}$ for regular cycles
- f4: Slope of the capacity fade curve between cycle 15 and 115
- f5: Slope of the capacity fade curve between cycle 90 and 115
- f6: Same as f1, but relativized (divided by cycles between)


***Notes***

- Could be more tidy to save this data as files and import them in a seperate script for modeling
    - Could do variance/mean etc in this cell and save them as raw features, then try ML with raw, log-transformed, sqrt-transformed etc
    - Alternatively: Gather all the data in one dataframe, with a column for each feature (and one for cell name)
- Actual cycles used as standard, but can switch out 'Cycle - total' with 'Cycle - total - normalized'
- To change groups, change the group variable. To use all cells, use find and replace and switch 'bat_dict[group]' with 'cell_dict'
- f6 seem to be the best feature when tested on all cells and the T25 group.


```python
cycle_lives = []

Qd_c20_fl = []
dQdV_c20_fl = []

Qd115_15 = []

slope1 = []
slope2 = []

R_DC = []

Qd_c20_fl_rel = []

group = 'T25'
cycle = 'Cycle - total'

for cell in bat_dict[group].keys():
    
    summary = bat_dict[group][cell]['summary']
    c20 = summary.loc[summary['Current_dch (C)'] == -0.05]
    c20_Qd = bat_dict[group][cell]['C20 discharge']
    c15 = bat_dict[group][cell]['c15']
    c115 = bat_dict[group][cell]['c115']
    
    # We define the cycle life as the cycle at which the capacity drops below 90% of the first measured c20 cycle capacity
    cl = c20[cycle][c20['Ah_cha (C20 first) (%)'] < 90].tolist()
    
    # Alternative definitions of cycle life
    #cl = c20['Ah_dch (C20 first) (%)'][c20['Cycle - total - normalized'] > 1000].tolist() # capacity at cycle 1000
    #cl = summary['Ah_dch (C20 first) (%)'][summary['Cycle - total - normalized'] > 1000].tolist()
    
    if(cl): # If the list is not empty aka cell does not go under 90%
        
        # f1
        
        # Find index of cycle in interval
        boolarr = c20[cycle].between(90, 220) # This varies between groups and must be tuned
        firstTrue = np.where(boolarr==True)
        ind = firstTrue[0][0]
        cycle_diff = c20.iloc[ind][cycle] - c20.iloc[0][cycle]
        
        
        c20_first_Qd = c20_Qd[0]['Ah_balance (normalized)'].values 
        c20_100_Qd = c20_Qd[ind]['Ah_balance (normalized)'].values
        

        
        # if different length, use only length of shortest array (should interpolate instead)
        if(len(c20_first_Qd) < len(c20_100_Qd)):
            c20_Qd_diff = c20_100_Qd[:len(c20_first_Qd)] - c20_first_Qd
            c20_Qd_diff_rel = (c20_100_Qd[:len(c20_first_Qd)] - c20_first_Qd)/cycle_diff
        elif(len(c20_first_Qd) > len(c20_100_Qd)):
            c20_Qd_diff = c20_100_Qd - c20_first_Qd[:len(c20_100_Qd)]
            c20_Qd_diff_rel = (c20_100_Qd - c20_first_Qd[:len(c20_100_Qd)])/cycle_diff
        else:
            c20_Qd_diff = c20_100_Qd - c20_first_Qd
            c20_Qd_diff_rel = (c20_100_Qd - c20_first_Qd)/cycle_diff
        
        
        # f2
        
        # dQdV C20
        dQdV_first = c20_Qd[0]['dQ/dV (1/V)'].values 
        dQdV_last = c20_Qd[ind]['dQ/dV (1/V)'].values
        
        if(len(dQdV_first) < len(dQdV_last)):
            dQdV_diff = dQdV_last[:len(dQdV_first)] - dQdV_first
        elif(len(dQdV_first) > len(dQdV_last)):
            dQdV_diff = dQdV_last - dQdV_first[:len(dQdV_last)]
        else:
            dQdV_diff = dQdV_last - dQdV_first
       
        
        
        # f3
        
        # Qd cycling
        Qd15 = c15['Ah_balance (normalized)'].values
        Qd115 = c115['Ah_balance (normalized)'].values
        
        if(len(Qd15) < len(Qd115)):
            Qd_diff = Qd115[:len(Qd15)] - Qd15
        elif(len(Qd15) > len(Qd115)): 
            Qd_diff = Qd115 - Qd15[:len(Qd115)]
        else:
            Qd_diff = Qd115 - Qd15
        
        
        
        
        # f5
        
        # slope of capacity fade
        Q115 = summary.loc[summary[cycle] == 100]['Ah_cha (C20 first) (%)'].values[0] # switch to > for normalized cycles
        Q15 = summary.loc[summary[cycle] == 15]['Ah_cha (C20 first) (%)'].values[0]
        Q90 = summary.loc[summary[cycle] == 90]['Ah_cha (C20 first) (%)'].values[0]
        
        slope15_115 = float((Q115 - Q15)/100)
        slope90_115 = float((Q115 - Q90)/25)
        
        
        # f6
        ir_15 = summary.loc[summary[cycle] > 15]['R_DC 50ms Soc50 cha (mOhm)'].values[0]
        ir_115 = summary.loc[summary[cycle] > 115]['R_DC 50ms Soc50 cha (mOhm)'].values[0]
        
        ir_diff = ir_115 - ir_15
        
        # Add to lists
        cycle_lives.append(cl[0])

        Qd_c20_fl.append(c20_Qd_diff)
        dQdV_c20_fl.append(dQdV_diff)
        
        Qd115_15.append(Qd_diff)
        
        slope1.append(slope15_115)
        slope2.append(slope90_115)
        
        R_DC.append(ir_diff)
        
        Qd_c20_fl_rel.append(c20_Qd_diff_rel)
        
print(len(cycle_lives), len(Qd_c20_fl), len(dQdV_c20_fl), len(Qd115_15), len(slope1), len(slope2))
```

### A simple Elastic Net model

```python
# Features from Qd_c20_fl
X = np.zeros((len(cycle_lives), 11)) # samples x features
y = np.zeros(len(cycle_lives)) # samples

for i in range(len(Qd_c20_fl)):
    # cycle life
    y[i] = cycle_lives[i]
    
    # C20 features
    X[i,0] = float(np.var(Qd_c20_fl[i])) # Variance
    X[i,1] = abs(np.mean(Qd_c20_fl[i])) # Mean
    X[i,2] = float(np.var(dQdV_c20_fl[i])) # Variance
    X[i,3] = np.mean(dQdV_c20_fl[i]) # Mean: Equivalent to the capacity
    
    # Cycling features
    X[i,4] = float(np.var(Qd115_15[i])) # Variance
    X[i,5] = np.mean(Qd115_15[i]) # Mean
    
    # Summary
    X[i,6] = slope1[i]
    X[i,7] = slope2[i]
    X[i,8] = R_DC[i]
    
    X[i,9] = float(np.var(Qd_c20_fl_rel[i])) # var
    X[i,10] = abs(np.mean(Qd_c20_fl_rel[i])) # Mean
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

errors = np.zeros((8, 20))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
ypred_train_arr = np.zeros((20, len(y_train)))
ypred_test_arr = np.zeros((20, len(y_test)))

kfold = KFold(n_splits = 5, shuffle=True, random_state=22)

# Fit model and make predictions 20 times for consistent results
for i in range(20):
    
    # Train/test split
        
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    # Elastic net pipeline
    alphas = [0.1,0.3,0.5,0.7,0.9,0.95,0.99,1]
    pipe_EN = Pipeline([('scaler', StandardScaler()), ('estimator', ElasticNetCV(cv=5, l1_ratio=alphas, max_iter=10000, tol=0.01))])
    print(X_train.shape)
    print(y_train.shape)
    pipe_EN.fit(X_train, y_train)
    ypred_train = pipe_EN.predict(X_train)
    ypred_test = pipe_EN.predict(X_test)

    # For calulating the mean predictions
    ypred_train_arr[i,:] = ypred_train
    ypred_test_arr[i,:] = ypred_test
    
    # Compute the mean for referance to model performance
    ypred_ref_train = np.mean(y_train)
    ypred_ref_test = np.mean(y_test)
    
    
    enet = ElasticNetCV(cv=5, l1_ratio=alphas, max_iter=10000, tol=0.01)
    print(cross_val_score(enet, X, y, cv=5))

    # Get errors
    errors[:,i] = get_errors2(y_train, y_test, ypred_train, ypred_test)
    print('RMSE train: ', errors[2,i])
    print('RMSE test: ', errors[3,i])
    print('MAPE train: ', round(100*errors[-2, i]))
    print('MAPE train: ', round(100*errors[-1, i]))
    print('')

errors = np.mean(errors, axis=1)    
errors_table = pd.DataFrame({'MSE': [errors[0], errors[1]], 'RMSE CYCLES': [errors[2],errors[3]],\
                       'R2 SCORE': [errors[4],errors[5]] ,'MAPE': [str(round(100*errors[6])) + ' %' , str(round(100*errors[7])) + ' %']}, index=['train', 'test'])


display(errors_table)
```
