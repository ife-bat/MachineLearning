# ML on IFE data

**In this notebook:**
- A dictionary is created to make it easy to access all data
- A simple Elastic Net model is created
    - Only C20 cycles included for each cell
    - Cycle life defined as the normalized cycle number when normalized capacity reaches 90% (first c20 cycle when this occurs)
    - Features are generated from the difference between the first and last C20 cycle, common name c20_diff
        - variance and mean of Qd_c20_diff, T_c20_diff, dQdV_c20_diff, total 6 features
    - RMSE cycles using only var(Qd_c20_diff): train: 300, test: 283
    - RMSE cycles using all mentioned features: train: 248, test: 298
    - RMSE cycles using all features and regular cycle life: train: 593, test: 731

    - Include more features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import os, random
import pathlib
import glob

pd.set_option('display.max_columns', 100)

from sklearn.pipeline import Pipeline
from sklearn.linear_model import ElasticNetCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold
from functions import get_errors
from sklearn.metrics import mean_squared_error, r2_score
```

### Make cell dictionary

```python
# Make a nested dictionary to easily access all data. first keys are cell-ids, second keys are charge, discharge or summary data
# Only c20 cycles are included at this point
cell_dict = {}

path_cells = r"C:\Users\ife12216\OneDrive - Institutt for Energiteknikk\Documents\Masteroppgave\ML_github\data\IFE\interim"

# Loop through cells
for cell in os.listdir(path_cells)[1:]: # Exclude Cell_ID_MSB19108E0
    
    # Make summary dataframes
    cell_dir = os.path.join(path_cells, cell)
    summary_file = os.path.join(cell_dir, os.listdir(cell_dir)[1])
    summary_df = pd.read_csv(summary_file, encoding = "ISO-8859-1", sep='\t')
    
    charge_cycles = []
    discharge_cycles = []
    cell_dict[cell] = {} # each cell is a new dictionary
    
    c20_dir = os.path.join(path_cells, cell+'\C20_cycles')
    
    # Loop through files with charge cycles and add them to list
    for charge_cycle in glob.glob(os.path.join(c20_dir, '*cha.txt')):
        charge_cycle_file = os.path.join(c20_dir, charge_cycle)
        charge_cycle_df = pd.read_csv(charge_cycle_file, encoding = "ISO-8859-1", sep='\t')
        charge_cycles.append(charge_cycle_df)
    
    # Loop through files with discharge cycles and add them to list
    for discharge_cycle in glob.glob(os.path.join(c20_dir, '*dch.txt')):
        discharge_cycle_file = os.path.join(c20_dir, discharge_cycle)
        discharge_cycle_df = pd.read_csv(discharge_cycle_file, encoding = "ISO-8859-1", sep='\t')
        discharge_cycles.append(discharge_cycle_df)

    # Add summary, charge cycles and discharge cycles to cell dictionary
    cell_dict[cell]['summary'] = summary_df
    cell_dict[cell]['charge'] = charge_cycles
    cell_dict[cell]['discharge'] = discharge_cycles
```

```python
cycle_lives = []

Qd_c20_fl = []
T_c20_fl = []
dQdV_c20_fl = []


for cell in cell_dict.keys():
    summary = cell_dict[cell]['summary']
    c20 = summary.loc[summary['Current_dch (C)'] == -0.05]
    cl = c20['Cycle - total - normalized'][c20['Ah_dch (C20 first) (%)'] < 90].tolist()
    if(cl): # If the list is not empty aka cell does not go under 90%
        
        # Qd
        Qd_first = cell_dict[cell]['discharge'][0]['Ah_balance (normalized)'].values[:1159] # Index 1159 because length varies, have to pick the shortest one
        Qd_last = cell_dict[cell]['discharge'][-1]['Ah_balance (normalized)'].values[:1159]
        Qd_diff = Qd_last - Qd_first
        
        # T
        T_first = cell_dict[cell]['discharge'][0]['Temperature (°C)'].values[:1159]
        T_last = cell_dict[cell]['discharge'][-1]['Temperature (°C)'].values[:1159]
        T_diff = T_last - T_first
        
        # dQdV
        dQdV_first = cell_dict[cell]['discharge'][0]['dQ/dV (1/V)'].values[:1159]
        dQdV_last = cell_dict[cell]['discharge'][-1]['dQ/dV (1/V)'].values[:1159]
        dQdV_diff = dQdV_last - dQdV_first
       
        # Add to lists
        cycle_lives.append(cl[0])

        Qd_c20_fl.append(Qd_diff)
        T_c20_fl.append(T_diff)
        dQdV_c20_fl.append(dQdV_diff)
    
print(len(cycle_lives), len(Qd_c20_fl), len(T_c20_fl), len(dQdV_c20_fl))
```

```python
cycle_lives
```

### A simple Elastic Net model

```python
# Features from Qd_c20_fl
X = np.zeros((len(cycle_lives), 6)) # samples x features
y = np.zeros(len(cycle_lives)) # samples

for i in range(len(Qd_c20_fl)):
    y[i] = np.log10(cycle_lives[i])
    X[i,0] = np.log10(float(np.var(Qd_c20_fl[i]))) # Variance
    X[i,1] = np.log10(abs(np.mean(Qd_c20_fl[i]))) # Mean
    X[i,2] = np.log10(float(np.var(T_c20_fl[i]))) # Variance
    X[i,3] = np.log10(abs(np.mean(T_c20_fl[i]))) # Mean
    X[i,4] = np.log10(float(np.var(dQdV_c20_fl[i]))) # Variance
    X[i,5] = np.log10(abs(np.mean(dQdV_c20_fl[i]))) # Mean
```

#### Visualize correlation between features and cycle lives

```python
plt.figure(figsize=(10,7))
plt.scatter(np.power(10,X[:,0]), np.power(10,y), c=y, cmap='RdBu')
plt.xscale('log')
plt.yscale('log')
plt.xticks([1e-3, 1e-2])
plt.xlabel('var(Qd_diff)')
plt.ylabel('cycle life')
plt.title('Qd var')
plt.show()

plt.figure(figsize=(10,7))
plt.scatter(np.power(10,X[:,1]), np.power(10,y), c=y, cmap='RdBu')
plt.xscale('log')
plt.yscale('log')
plt.xticks()
plt.xlabel('mean(Qd_diff)')
plt.ylabel('cycle life')
plt.title('Qd mean')
plt.show()

plt.figure(figsize=(10,7))
plt.scatter(np.power(10,X[:,2]), np.power(10,y), c=y, cmap='RdBu')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('var(T_diff)')
plt.ylabel('cycle life')
plt.title('T var')
plt.show()

plt.figure(figsize=(10,7))
plt.scatter(np.power(10,X[:,3]), np.power(10,y), c=y, cmap='RdBu')
plt.xscale('log')
plt.yscale('log')
plt.xticks()
plt.xlabel('mean(T_diff)')
plt.ylabel('cycle life')
plt.title('T mean')
plt.show()

plt.figure(figsize=(10,7))
plt.scatter(np.power(10,X[:,4]), np.power(10,y), c=y, cmap='RdBu')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('var(dQdV_diff)')
plt.ylabel('cycle life')
plt.title('dQdV var')
plt.show()

plt.figure(figsize=(10,7))
plt.scatter(np.power(10,X[:,5]), np.power(10,y), c=y, cmap='RdBu')
plt.xscale('log')
plt.yscale('log')
plt.xticks()
plt.xlabel('mean(dQdV_diff)')
plt.ylabel('cycle life')
plt.title('dQdV mean')
plt.show()
```

#### Make model and do predictions

```python
X_var = X[:,0].reshape(-1,1) # For only using variance feature


errors = np.zeros((8, 20))

# Fit model and make predictions 20 times for consistent results
for i in range(20):
    
    # Train/test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    # Elastic net pipeline
    alphas = [0.1,0.3,0.5,0.7,0.9,0.95,0.99,1]
    pipe_EN = Pipeline([('scaler', StandardScaler()), ('estimator', ElasticNetCV(cv=10, l1_ratio=alphas, max_iter=10000))])
    
    pipe_EN.fit(X_train, y_train)
    ypred_train = pipe_EN.predict(X_train)
    ypred_test = pipe_EN.predict(X_test)

    # Get errors
    errors[:,i] = get_errors(y_train, y_test, ypred_train, ypred_test)
    
errors = np.mean(errors, axis=1)    
errors_table = pd.DataFrame({'RMSE': [errors[0], errors[1]], 'RMSE CYCLES': [errors[2],errors[3]],\
                       'R2 SCORE': [errors[4],errors[5]] ,'MAPE': [errors[6],errors[7]]}, index=['train', 'test'])

display(errors_table)
```

```python
# Vekt av de 6 featurene
print(pipe_EN[1].coef_)
# Hver feature får en vekt. Vi ser på forskjell på discharge capacity. En gj.sn. stor forskjell korrelerer til dårlig levetid. 
# Hvis man da har forskjellige betingelser, kan man ha liten forskjell i Qd,
# men likevel kort levetid på grunn av en annen betingelse? Vekten vil da bli kunstig lav?

# Vi vil vel egentlig:
# En modell per syklingsbetingelse: forskjellige temp, c-rate, soc-vindu
# Får da forskjellige koeffisienter på hver modell
# Hva får man egentlig ut da, sammenlignet med å bare anta cycle life basert på erfaring fra tidligere cap fade kurver?
# Kan skje forskjellige ting (random kjemi) med like betingelser, ML vil fange opp det gjennom featurene. 
# Forskjellige betingelser: eks: kjemisk degradering kan skje ved 2% soc (tvinger litium ut av anoden). Celle med
# 50-70% soc vindu vil aldri få med seg dette, så vekten som blir gitt til en feature basert på denne cellen blir feil
# (hvis featuren påvirkes mye ved 2%)
```
