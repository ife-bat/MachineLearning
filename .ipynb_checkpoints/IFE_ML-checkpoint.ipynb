# ML on IFE data

**In this notebook:**
- A dictionary is created to make it easy to access all data
    - One for all cells and others for the subsets of cells
- A simple Elastic Net model is created
    - Cycle life defined as the normalized cycle number when normalized capacity reaches 90% (first c20 cycle when this occurs)
    - Features are generated from the difference between the first and last C20 cycle, common name c20_diff
        - variance and mean of Qd_c20_diff, dQdV_c20_diff, total 4 features
- Next step will be to

### Extract relevant data

```python
cycle_lives = []

Qd_c20_fl = []
dQdV_c20_fl = []
Qd115_15 = []
Qd115_15_norm = []
slope15_115 = []
slope90_115 = []
slope15_115_norm = []
slope90_115_norm = []

for cell in cell_dict.keys():
    summary = cell_dict[cell]['summary']
    c20 = summary.loc[summary['Current_dch (C)'] == -0.05]
    
    # cycle life definition
    cl = c20['Cycle - total - normalized'][c20['Ah_dch (C20 first) (%)'] < 90].tolist() # cap rel to first c20 cycle
    
    # Other definitions of cycle life
    #cl = c20['Ah_dch (C20 first) (%)'][c20['Cycle - total - normalized'] > 1000].tolist() # capacity at cycle 1000
    #cl = summary['Ah_dch (C20 first) (%)'][summary['Cycle - total - normalized'] > 1000].tolist()
    
    if(cl): # If the list is not empty aka cell does not go under 90%
        
        # Find c20 cycle between 100 and 201
        boolarr = c20['Cycle - total - normalized'].between(100, 201)
        firstTrue = np.where(boolarr==True) # first index where cycle is between 100 and 201
        ind = firstTrue[0][0]
        cycle_norm_diff = c20.iloc[ind]['Cycle - total - normalized'] - c20.iloc[0]['Cycle - total - normalized']
        
        # Qd C20
        c20_first_Qd = cell_dict[cell]['discharge'][0]['Ah_balance (normalized)'].values 
        c20_100_Qd = cell_dict[cell]['discharge'][ind]['Ah_balance (normalized)'].values
        
        # if different length, use only length of shortest array (should interpolate instead)
        if(len(c20_first_Qd) < len(c20_100_Qd)):
            c20_Qd_diff = c20_100_Qd[:len(c20_first_Qd)] - c20_first_Qd
        elif(len(c20_first_Qd) > len(c20_100_Qd)):
            c20_Qd_diff = c20_100_Qd - c20_first_Qd[:len(c20_100_Qd)]
        else:
            c20_Qd_diff = c20_100_Qd - c20_first_Qd
        
        
        # dQdV C20
        dQdV_first = cell_dict[cell]['discharge'][0]['dQ/dV (1/V)'].values
        dQdV_last = cell_dict[cell]['discharge'][ind]['dQ/dV (1/V)'].values
        
        if(len(dQdV_first) < len(dQdV_last)):
            dQdV_diff = dQdV_last[:len(dQdV_first)] - dQdV_first
        elif(len(dQdV_first) > len(dQdV_last)):
            dQdV_diff = dQdV_last - dQdV_first[:len(dQdV_last)]
        else:
            dQdV_diff = dQdV_last - dQdV_first
       
        
        # Qd cycling
        Qd15 = cell_dict[cell]['c15']['Ah_balance (mAh)'].values
        Qd115 = cell_dict[cell]['c115']['Ah_balance (mAh)'].values
        
        if(len(Qd15) < len(Qd115)):
            Qd_diff = Qd115[:len(Qd15)] - Qd15
        elif(len(Qd15) > len(Qd115)): 
            Qd_diff = Qd115 - Qd15[:len(Qd115)]
        else:
            Qd_diff = Qd115 - Qd15
            
        Qd15_norm = cell_dict[cell]['c15 normalized']['Ah_balance (mAh)'].values
        Qd115_norm = cell_dict[cell]['c115 normalized']['Ah_balance (mAh)'].values
        
        if(len(Qd15_norm) < len(Qd115_norm)):
            Qd_diff_norm = Qd115_norm[:len(Qd15_norm)] - Qd15_norm
        elif(len(Qd15_norm) > len(Qd115_norm)): 
            Qd_diff_norm = Qd115_norm - Qd15_norm[:len(Qd115_norm)]
        else:
            Qd_diff_norm = Qd115_norm - Qd15_norm
        
        # slope of capacity fade
        Q115 = cell_dict[cell]['summary'].loc[cell_dict[cell]['summary']['Cycle - total'] == 100]['Ah_dch (C20 first) (%)'].values[0]
        Q15 = cell_dict[cell]['summary'].loc[cell_dict[cell]['summary']['Cycle - total'] == 15]['Ah_dch (C20 first) (%)'].values[0]
        Q90 = cell_dict[cell]['summary'].loc[cell_dict[cell]['summary']['Cycle - total'] == 90]['Ah_dch (C20 first) (%)'].values[0]
        
        Q115_norm = cell_dict[cell]['summary'].loc[cell_dict[cell]['summary']['Cycle - total - normalized'] > 100]['Ah_dch (C20 first) (%)'].values[0]
        Q15_norm = cell_dict[cell]['summary'].loc[cell_dict[cell]['summary']['Cycle - total - normalized'] > 15]['Ah_dch (C20 first) (%)'].values[0]
        Q90_norm = cell_dict[cell]['summary'].loc[cell_dict[cell]['summary']['Cycle - total - normalized'] > 90]['Ah_dch (C20 first) (%)'].values[0]
        
        # Add to lists
        cycle_lives.append(cl[0])

        Qd_c20_fl.append(c20_Qd_diff)
        
        dQdV_c20_fl.append(dQdV_diff)
        
        Qd115_15.append(Qd_diff)
        Qd115_15_norm.append(Qd_diff_norm)
        
        slope15_115.append((Q115 - Q15)/100)
        slope90_115.append((Q115 - Q90)/25)
        slope15_115_norm.append((Q115_norm - Q15_norm)/100)
        slope90_115_norm.append((Q115_norm - Q90_norm)/25)
        
        #print(c20.iloc[ind]['Cycle - total - normalized'])
        
print(len(cycle_lives), len(Qd_c20_fl), len(dQdV_c20_fl), len(Qd115_15), len(Qd115_15_norm), len(slope15_115),\
      len(slope90_115), len(slope15_115_norm), len(slope90_115_norm))
```

### A simple Elastic Net model

```python
# Features from Qd_c20_fl
X = np.zeros((len(cycle_lives), 12)) # samples x features
y = np.zeros(len(cycle_lives)) # samples

for i in range(len(Qd_c20_fl)):
    # cycle life
    y[i] = np.log10(cycle_lives[i])
    
    # C20 features
    X[i,0] = np.log10(float(np.var(Qd_c20_fl[i]))) # Variance
    X[i,1] = np.log10(abs(np.mean(Qd_c20_fl[i]))) # Mean
    X[i,2] = np.log10(float(np.var(dQdV_c20_fl[i]))) # Variance
    X[i,3] = np.log10(abs(np.mean(dQdV_c20_fl[i]))) # Mean: Equivalent to the capacity
    
    # Cycling features
    X[i,4] = np.log10(float(np.var(Qd115_15[i]))) # Variance
    X[i,5] = np.log10(abs(np.mean(Qd115_15[i]))) # Mean
    X[i,6] = np.log10(float(np.var(Qd115_15_norm[i]))) # Variance
    X[i,7] = np.log10(abs(np.mean(Qd115_15_norm[i]))) # Mean
    
    # Summary
    X[i,8] = slope15_115[i]
    X[i,9] = slope90_115[i]
    X[i,10] = slope15_115_norm[i]
    X[i,11] = slope90_115_norm[i]
    
```

#### Make model and do predictions

```python
X_var = X[:,0].reshape(-1,1) # For only using variance feature


errors = np.zeros((8, 20))

# Fit model and make predictions 20 times for consistent results
for i in range(20):
    
    # Train/test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    # Elastic net pipeline
    alphas = [0.1,0.3,0.5,0.7,0.9,0.95,0.99,1]
    pipe_EN = Pipeline([('scaler', StandardScaler()), ('estimator', ElasticNetCV(cv=5, l1_ratio=alphas, max_iter=10000, tol=0.001))])
    
    pipe_EN.fit(X_train, y_train)
    ypred_train = pipe_EN.predict(X_train)
    ypred_test = pipe_EN.predict(X_test)

    # Get errors
    errors[:,i] = get_errors(y_train, y_test, ypred_train, ypred_test)

errors = np.mean(errors, axis=1)    
errors_table = pd.DataFrame({'RMSE': [errors[0], errors[1]], 'RMSE CYCLES': [errors[2],errors[3]],\
                       'R2 SCORE': [errors[4],errors[5]] ,'MAPE': [errors[6],errors[7]]}, index=['train', 'test'])

display(errors_table)
```
